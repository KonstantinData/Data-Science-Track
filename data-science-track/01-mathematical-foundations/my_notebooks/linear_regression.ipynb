```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Daten vorbereiten
y = np.array([632000, 554000, 310000, 506000, 422000, 658000, 634000, 720000, 408000, 500000])
x = np.array([[185], [197], [117], [155], [145], [268], [225], [276], [132], [199]])  # 2D-Array für sklearn

# Modell erstellen und trainieren
reg = LinearRegression()
reg.fit(x, y)

# Vorhersage für einen neuen Wert (mit Benutzereingabe)
square_meters = float(input("Gib die Quadratmeterzahl ein: "))  # Eingabe als Float umwandeln
new_x = np.array([[square_meters]])  # In 2D-Array umwandeln, da sklearn das erwartet
predicted_price = reg.predict(new_x)

print(f"Der vorhergesagte Preis für {square_meters} m²: {predicted_price[0]:,.2f} EUR")

# Visualisierung
plt.scatter(x, y, color="blue", label="Datenpunkte")
plt.plot(x, reg.predict(x), color="red", label="Regressionslinie")
plt.scatter(new_x, predicted_price, color="green", label="Vorhersagewert", marker="x", s=100)  # Grünes X für neue Vorhersage
plt.xlabel("Square Meters")
plt.ylabel("House Price")
plt.legend()
plt.show()
```
