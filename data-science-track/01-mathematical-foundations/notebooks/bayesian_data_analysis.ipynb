### Explanation KDE Plot

A **KDE plot (Kernel Density Estimate plot)** is a method for estimating the probability density function of a continuous random variable. It provides a smooth representation of data distribution, similar to a histogram but without relying on discrete bins.

##### Characteristics of a KDE Plot:
- **Smoothing the Distribution**: Unlike a histogram, which uses bars, a KDE plot uses a smooth curve to estimate the density of data points.
- **Kernel Function**: A mathematical function (e.g., Gaussian kernel) is applied to smooth the data distribution.
- **Bandwidth Parameter**: Controls the level of smoothing. A small bandwidth captures more details, while a larger bandwidth results in a smoother curve.
- **Alternative to Histograms**: KDE plots provide a continuous estimate of the density, reducing the dependency on bin sizes used in histograms.

##### Use Cases:
- Understanding the distribution of data
- Comparing multiple density distributions
- Visualizing probability densities

### Probability Rules

**Sum rule**
- Probability of A or B (independent events)
- OR = addition
- Probability of rolling 2 or 4 with a die

$$
P(2 \text{ oder } 4) = P(2) + P(4) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3} \approx 33.3\%
$$


**Product rule**
- Probability of A and B (independent events)
- AND = multiplication
- Probability of rolling 2 and then 4 with a die

$$
P(2 \text{ und dann } 4) = P(2) \cdot P(4) = \frac{1}{6} \times \frac{1}{6} = \frac{1}{36} \approx 2.8\%
$$

# Conditional probability

- Probability of event A given that event B has occurred
- Definition: 

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}, \quad \text{wenn } P(B) > 0
$$

- Example: Probability of drawing an Ace given that a red card was drawn (assuming a standard deck of 52 cards)


# Bayes Theorem

**Bayes' Theorem**
- Describes the probability of event A given that event B has occurred
- Formula:

$$
P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}, \quad \text{wenn } P(B) > 0
$$

- Where:
  - \( P(A \mid B) \) = Posterior probability (probability of A given B)
  - \( P(B \mid A) \) = Likelihood (probability of B given A)
  - \( P(A) \) = Prior probability (initial probability of A)
  - \( P(B) \) = Normalizing constant (total probability of B)

- Example: Diagnosing a disease given a positive test result
