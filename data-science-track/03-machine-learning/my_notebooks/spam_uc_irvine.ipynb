```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler

# Daten laden
file_path = r"D:\UC_Irvine_Machine_Learning_Data\SPAM\spambase.data"
column_names = [f'feature_{i}' for i in range(57)] + ['is_spam']
spam_df = pd.read_csv(file_path, header=None, names=column_names)

# Fehlende Werte entfernen (falls vorhanden)
spam_df = spam_df.dropna()

# Features und Zielvariable definieren
X = spam_df.drop('is_spam', axis=1)
y = spam_df['is_spam']

# Daten in Trainings- und Testset aufteilen
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardisierung der Daten
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Random Forest Modell initialisieren und trainieren
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)

# Vorhersagen treffen
y_pred = model.predict(X_test_scaled)

# Modellbewertung
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification = classification_report(y_test, y_pred)

print(f"Accuracy Score: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report: {classification}")

# Confusion Matrix visualisieren
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Not Spam", "Spam"], yticklabels=["Not Spam", "Spam"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix Heatmap - Random Forest")
plt.show()

# Feature Importance visualisieren
feature_importances = model.feature_importances_
features = X.columns

plt.figure(figsize=(10, 5))
sns.barplot(x=features, y=feature_importances)
plt.xticks(rotation=90)
plt.xlabel("Features")
plt.ylabel("Feature Importance")
plt.title("Feature Importance - Random Forest")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler

# Daten laden
file_path = r"D:\UC_Irvine_Machine_Learning_Data\SPAM\spambase.data"
column_names = [f'feature_{i}' for i in range(57)] + ['is_spam']
spam_df = pd.read_csv(file_path, header=None, names=column_names)

# Fehlende Werte entfernen (falls vorhanden)
spam_df = spam_df.dropna()

# Features und Zielvariable definieren
X = spam_df.drop('is_spam', axis=1)
y = spam_df['is_spam']

# Daten in Trainings- und Testset aufteilen
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardisierung der Daten
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Logistic Regression Modell initialisieren und trainieren
model = LogisticRegression(max_iter=500, solver='lbfgs', random_state=42)
model.fit(X_train_scaled, y_train)

# Vorhersagen treffen
y_pred = model.predict(X_test_scaled)

# Modellbewertung
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification = classification_report(y_test, y_pred)

print(f"Accuracy Score: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report: {classification}")

# Confusion Matrix visualisieren
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Not Spam", "Spam"], yticklabels=["Not Spam", "Spam"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix Heatmap - Logistic Regression")
plt.show()

# Feature Importance (absoluter Wert der Koeffizienten)
feature_importances = np.abs(model.coef_)[0]
features = X.columns

# Visualisierung der Feature-Wichtigkeit
plt.figure(figsize=(10, 5))
sns.barplot(x=features, y=feature_importances)
plt.xticks(rotation=90)
plt.xlabel("Features")
plt.ylabel("Feature Importance (Abs Coefficients)")
plt.title("Feature Importance - Logistic Regression")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import StandardScaler

# Daten laden
file_path = r"D:\UC_Irvine_Machine_Learning_Data\SPAM\spambase.data"
column_names = [f'feature_{i}' for i in range(57)] + ['is_spam']
spam_df = pd.read_csv(file_path, header=None, names=column_names)

# Fehlende Werte entfernen (falls vorhanden)
spam_df = spam_df.dropna()

# Features und Zielvariable definieren
X = spam_df.drop('is_spam', axis=1)
y = spam_df['is_spam']

# Daten in Trainings- und Testset aufteilen
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardisierung der Daten
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# KNN-Modell initialisieren und trainieren
model = KNeighborsClassifier(n_neighbors=7, weights="distance", metric="cosine", algorithm="auto")
model.fit(X_train_scaled, y_train)

# Vorhersagen treffen
y_pred = model.predict(X_test_scaled)

# Modellbewertung
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification = classification_report(y_test, y_pred)

print(f"Accuracy Score: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report: {classification}")

# **1. Verteilung der Zielvariablen (Spam vs. Nicht-Spam)**
plt.figure(figsize=(6, 4))
sns.countplot(x=y)
plt.title("Verteilung der Spam-Klassen")
plt.xlabel("Klasse (0 = Nicht-Spam, 1 = Spam)")
plt.ylabel("Anzahl")
plt.show()

# **2. Confusion Matrix visualisieren**
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Not Spam", "Spam"], yticklabels=["Not Spam", "Spam"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix Heatmap - KNN")
plt.show()

# **3. Optimierung des k-Werts für KNN**
errors = []
neighbors = range(1, 21)

for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k, weights="distance", metric="cosine")
    knn.fit(X_train_scaled, y_train)
    y_pred_k = knn.predict(X_test_scaled)
    errors.append(1 - accuracy_score(y_test, y_pred_k))

plt.figure(figsize=(8, 5))
plt.plot(neighbors, errors, marker='o', linestyle='dashed', color='b')
plt.xlabel("Anzahl der Nachbarn (k)")
plt.ylabel("Fehlerrate")
plt.title("Optimierung des k-Werts für KNN")
plt.show()

# **4. ROC-Kurve und AUC-Wert**
y_proba = model.predict_proba(X_test_scaled)[:, 1]  # Wahrscheinlichkeiten für die Klasse "Spam"
fpr, tpr, _ = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--')  # Zufallslinie
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - KNN")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import StandardScaler
from scipy.stats import entropy

# Daten laden
file_path = r"D:\UC_Irvine_Machine_Learning_Data\SPAM\spambase.data"
column_names = [f'feature_{i}' for i in range(57)] + ['is_spam']
spam_df = pd.read_csv(file_path, header=None, names=column_names)

# Fehlende Werte entfernen (falls vorhanden)
spam_df = spam_df.dropna()

# Features und Zielvariable definieren
X = spam_df.drop('is_spam', axis=1)
y = spam_df['is_spam']

# **1. Berechnung der Shannon-Entropie für jedes Feature**
feature_entropies = []
for column in X.columns:
    values, counts = np.unique(X[column], return_counts=True)
    probabilities = counts / len(X[column])
    entropie_feature = entropy(probabilities, base=2)
    feature_entropies.append(entropie_feature)

# Entropie-Werte in DataFrame speichern
feature_entropy_df = pd.DataFrame({'Feature': X.columns, 'Entropie': feature_entropies})
feature_entropy_df = feature_entropy_df.sort_values(by='Entropie', ascending=False)

# **2. Entfernen von Features mit niedriger Entropie (Schwellwert: 0.1)**
low_entropy_features = feature_entropy_df[feature_entropy_df['Entropie'] < 0.1]['Feature'].tolist()
X_reduced = X.drop(columns=low_entropy_features)

print(f"Entfernte Features aufgrund niedriger Entropie: {low_entropy_features}")
print(f"Neue Anzahl an Features nach Filterung: {X_reduced.shape[1]}")

# **3. Modelltraining mit gefilterten Features**
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train_reduced, X_test_reduced, _, _ = train_test_split(X_reduced, y, test_size=0.3, random_state=42)

# Standardisierung der Daten
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_reduced_scaled = scaler.fit_transform(X_train_reduced)
X_test_reduced_scaled = scaler.transform(X_test_reduced)

# KNN-Modell für Originaldaten
model_full = KNeighborsClassifier(n_neighbors=7, weights="distance", metric="cosine")
model_full.fit(X_train_scaled, y_train)
y_pred_full = model_full.predict(X_test_scaled)

# KNN-Modell für gefilterte Daten
model_reduced = KNeighborsClassifier(n_neighbors=7, weights="distance", metric="cosine")
model_reduced.fit(X_train_reduced_scaled, y_train)
y_pred_reduced = model_reduced.predict(X_test_reduced_scaled)

# **4. Modellbewertung**
accuracy_full = accuracy_score(y_test, y_pred_full)
accuracy_reduced = accuracy_score(y_test, y_pred_reduced)

print(f"Accuracy mit allen Features: {accuracy_full:.4f}")
print(f"Accuracy mit Entropie-gefilterten Features: {accuracy_reduced:.4f}")

# **5. Visualisierung der Shannon-Entropie (Top 20 Features)**
plt.figure(figsize=(12, 5))
sns.barplot(x=feature_entropy_df['Feature'][:20], y=feature_entropy_df['Entropie'][:20], color='blue')
plt.xticks(rotation=90)
plt.xlabel("Feature")
plt.ylabel("Entropie")
plt.title("Top 20 Features nach Entropie")
plt.show()

# **6. Confusion Matrix für das Modell mit gefilterten Features**
conf_matrix_reduced = confusion_matrix(y_test, y_pred_reduced)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix_reduced, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Not Spam", "Spam"], yticklabels=["Not Spam", "Spam"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Entropie-gefiltertes KNN")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from scipy.stats import entropy

# Daten laden
file_path = r"D:\UC_Irvine_Machine_Learning_Data\SPAM\spambase.data"
column_names = [f'feature_{i}' for i in range(57)] + ['is_spam']
spam_df = pd.read_csv(file_path, header=None, names=column_names)

# Fehlende Werte entfernen (falls vorhanden)
spam_df = spam_df.dropna()

# Features und Zielvariable definieren
X = spam_df.drop('is_spam', axis=1)
y = spam_df['is_spam']

# **1. Berechnung der Shannon-Entropie für jedes Feature**
feature_entropies = []
for column in X.columns:
    values, counts = np.unique(X[column], return_counts=True)
    probabilities = counts / len(X[column])
    entropie_feature = entropy(probabilities, base=2)
    feature_entropies.append(entropie_feature)

# Entropie-Werte in DataFrame speichern
feature_entropy_df = pd.DataFrame({'Feature': X.columns, 'Entropie': feature_entropies})
feature_entropy_df = feature_entropy_df.sort_values(by='Entropie', ascending=False)

# **2. Entfernen von Features mit niedriger Entropie (Schwellwert: 0.1)**
low_entropy_features = feature_entropy_df[feature_entropy_df['Entropie'] < 0.1]['Feature'].tolist()
X_reduced = X.drop(columns=low_entropy_features)

print(f"Entfernte Features aufgrund niedriger Entropie: {low_entropy_features}")
print(f"Neue Anzahl an Features nach Filterung: {X_reduced.shape[1]}")

# **3. Feature Engineering: Normalisierung der Top 10 Features mit höchster Entropie**
top_entropy_features = feature_entropy_df['Feature'][:10].tolist()
scaler_minmax = MinMaxScaler()
X[top_entropy_features] = scaler_minmax.fit_transform(X[top_entropy_features])

# **4. Modelltraining mit gefilterten Features**
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train_reduced, X_test_reduced, _, _ = train_test_split(X_reduced, y, test_size=0.3, random_state=42)

# Standardisierung der Daten
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_reduced_scaled = scaler.fit_transform(X_train_reduced)
X_test_reduced_scaled = scaler.transform(X_test_reduced)

# KNN-Modell für Originaldaten
model_full = KNeighborsClassifier(n_neighbors=7, weights="distance", metric="cosine")
model_full.fit(X_train_scaled, y_train)
y_pred_full = model_full.predict(X_test_scaled)

# KNN-Modell für gefilterte Daten
model_reduced = KNeighborsClassifier(n_neighbors=7, weights="distance", metric="cosine")
model_reduced.fit(X_train_reduced_scaled, y_train)
y_pred_reduced = model_reduced.predict(X_test_reduced_scaled)

# **5. Modellbewertung**
accuracy_full = accuracy_score(y_test, y_pred_full)
accuracy_reduced = accuracy_score(y_test, y_pred_reduced)

print(f"Accuracy mit allen Features: {accuracy_full:.4f}")
print(f"Accuracy mit Entropie-gefilterten Features: {accuracy_reduced:.4f}")

# **6. Visualisierung der Shannon-Entropie (Top 20 Features)**
plt.figure(figsize=(12, 5))
sns.barplot(x=feature_entropy_df['Feature'][:20], y=feature_entropy_df['Entropie'][:20], color='blue')
plt.xticks(rotation=90)
plt.xlabel("Feature")
plt.ylabel("Entropie")
plt.title("Top 20 Features nach Entropie")
plt.show()

# **7. Confusion Matrix für das Modell mit gefilterten Features**
conf_matrix_reduced = confusion_matrix(y_test, y_pred_reduced)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix_reduced, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Not Spam", "Spam"], yticklabels=["Not Spam", "Spam"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Entropie-gefiltertes KNN")
plt.show()

# **8. ROC-Kurve für beide Modelle**
y_proba_full = model_full.predict_proba(X_test_scaled)[:, 1]
y_proba_reduced = model_reduced.predict_proba(X_test_reduced_scaled)[:, 1]

fpr_full, tpr_full, _ = roc_curve(y_test, y_proba_full)
fpr_reduced, tpr_reduced, _ = roc_curve(y_test, y_proba_reduced)

roc_auc_full = auc(fpr_full, tpr_full)
roc_auc_reduced = auc(fpr_reduced, tpr_reduced)

plt.figure(figsize=(8, 5))
plt.plot(fpr_full, tpr_full, label=f"Original Model (AUC = {roc_auc_full:.2f})", linestyle="--")
plt.plot(fpr_reduced, tpr_reduced, label=f"Reduced Model (AUC = {roc_auc_reduced:.2f})", linestyle="-")
plt.plot([0, 1], [0, 1], 'k--')  # Zufallslinie
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - KNN")
plt.legend()
plt.show()
```
