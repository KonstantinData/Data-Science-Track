Uniform distribution $U(a,b)$
===
PDF
---
$$f(x) = \begin{cases}
\frac{1}{b-a} & \text{for } x\in [a,b] \\
0 & \text{otherwise}
\end{cases}
$$

CDF
---
$$F(x) = \begin{cases}
0 & \text{for } x < a \\
\frac{x-a}{b-a} & \text{for } x\in [a,b] \\
1 & \text{for } x>b
\end{cases}
$$

Mean
---
$$\mu = \frac{1}{2}(a+b)$$

Variance
---
$$\sigma^2 = \frac{1}{12}(b-a)^2$$

```python
# write a code for uniform dist
def pdf(x,a,b):
    if x<a or x>b:
        return 0
    return 1/(b-a)

def cdf(x,a,b):
    if x<a:
        return 0
    if x>b:
        return 1
    return (x-a)/(b-a)
```

```python
#make plots
from plotly.graph_objects import *
import numpy as np

x = np.linspace(-10,15,1000)
a, b = 1, 5

fig = Figure(data=Scatter(x = x,
                          y = [pdf(i,a,b) for i in x],
                          name = 'PDF',
                          showlegend=True))
fig.show()

fig = Figure(data=Scatter(x = x,
                          y = [cdf(i, a, b) for i in x],
                          name = 'CDF',
                          showlegend=True))
fig.show()
```

Exponential distribution $\text{Exp}(\lambda)$
===
PDF
--
$$f(x) = \lambda e^{-\lambda x}$$

CDF
---
$$F(x) = 1-e^{-\lambda x}$$


Mean
--
$$\mu = \frac{1}{\lambda}$$

Variance
---
$$\sigma^2 = \frac{1}{\lambda^2}$$

```python
# plot pdf and cdf
from scipy.stats import expon

x = np.linspace(0,20)
ls = [0.5, 1, 1.5]

traces_pdf = list()
traces_cdf = list()
for l in ls:
  traces_pdf.append(Scatter(x = x,
                        y = expon.pdf(x, scale = 1/l),
                        name = str(l)))
  traces_cdf.append(Scatter(x = x,
                        y = expon.cdf(x, scale = 1/l),
                        name = str(l)))
fig = Figure(data = traces_pdf,
             layout = Layout(title = "PDF"))
fig.show()
fig = Figure(data = traces_cdf,
             layout = Layout(title = "CDF"))
fig.show()
```
